```python
import csv
import ads
import pandas as pd
```

```python
data_frame = pd.read_csv('mars_gaz.csv', index_col = False)
```

```python
attr = data_frame[['Feature_ID', 'Clean_Feature_Name','Feature_Type', 'Feature_Type_Code']]
```

```python
cno = 0
Papers = []
for i, j in zip(attr['Clean_Feature_Name'], range(len(attr))):
    q = ads.SearchQuery(q=i, keyword='mars', fl=['bibcode', 'title', 'author', 'year', 'pub', 'ciation_count'], rows = 2000)
    fid = attr['Feature_ID'][j] 
    cfn = attr['Clean_Feature_Name'][j] 
    ft = attr['Feature_Type'][j] 
    ftc = attr['Feature_Type_Code'][j]
    for p in q:
        article = (fid, cfn, ft, ftc, p.title, p.author, p.year, p.bibcode, p.pub, p.citation_count)
        Papers.append(article)
    cno += 1
    print(str(cno) + i + " info added ")
```

```python
dfpaper = pd.DataFrame(Papers)
dfpaper.columns = ['Feature_ID', 'Clean_Feature_Name','Feature_Type', 'Feature_Type_Code', 'title', 'author', 'year', 'bibcode', 'pub', 'number of citations']
```

```python
dfpaper.to_csv("paper.csv", sep=',', header=True, index=False, encoding='utf-8')
```
